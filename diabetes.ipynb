```python
import numpy as np

# Replace 'merged_df' with your actual DataFrame containing the columns % INACTIVE, % OBESE, and % DIABETIC
inactive_data = merged_df['% INACTIVE'].values
obese_data = merged_df['% OBESE'].values
diabetic_data = merged_df['% DIABETIC'].values
# Number of bootstrap samples
num_samples = 1000  # You can adjust this number based on your needs

# Function to calculate the mean and confidence interval
def calculate_bootstrap_ci(data):
    bootstrap_means = []
    for _ in range(num_samples):
        # Sample with replacement from the data
        bootstrap_sample = np.random.choice(data, size=len(data), replace=True)
        # Calculate the mean of the bootstrap sample and store it
        bootstrap_means.append(np.mean(bootstrap_sample))
    
    # Calculate the confidence interval (e.g., 95% CI)
    lower_bound = np.percentile(bootstrap_means, 2.5)
    upper_bound = np.percentile(bootstrap_means, 97.5)
    
    return lower_bound, upper_bound

# Calculate bootstrap confidence intervals for each variable
inactive_lower, inactive_upper = calculate_bootstrap_ci(inactive_data)
obese_lower, obese_upper = calculate_bootstrap_ci(obese_data)
diabetic_lower, diabetic_upper = calculate_bootstrap_ci(diabetic_data)

# Print the results
print(f'Bootstrap 95% Confidence Interval for % INACTIVE: ({inactive_lower:.2f}, {inactive_upper:.2f})')
print(f'Bootstrap 95% Confidence Interval for % OBESE: ({obese_lower:.2f}, {obese_upper:.2f})')
print(f'Bootstrap 95% Confidence Interval for % DIABETIC: ({diabetic_lower:.2f}, {diabetic_upper:.2f})')
```
